## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "./data/Mouse_metadata.csv"
study_results_path = "./data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results,
                       on='Mouse ID',
                       how='inner')

# Display the data table for preview
combined_df = combined_df.rename(columns={'Tumor Volume (mm3)': "Tumor Volume", 'Weight (g)': 'Weight'})
combined_df
```

```python
# Checking the number of mice.
total_mice = len(combined_df['Mouse ID'])
total_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. CHECK
mice_duplicates = combined_df[combined_df.duplicated(['Mouse ID', 'Timepoint'])]
mice_duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicates = combined_df[combined_df.duplicated(['Mouse ID'])]
duplicates
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.c
clean_df = combined_df.drop_duplicates('Mouse ID')
    
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
total_mice_clean = len(clean_df['Mouse ID'])
total_mice_clean
clean_df = clean_df.rename(columns={'Tumor Volume (mm3)': "Tumor Volume"})
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

mean = combined_df.groupby('Drug Regimen')['Tumor Volume'].mean().round(2)
median = combined_df.groupby('Drug Regimen')['Tumor Volume'].mean().round(2)
variance =  combined_df.groupby('Drug Regimen')['Tumor Volume'].var().round(2)
std_dev = combined_df.groupby('Drug Regimen')['Tumor Volume'].std().round(2)
sem = combined_df.groupby('Drug Regimen')['Tumor Volume'].sem().round(2)
regimen = combined_df.groupby('Drug Regimen')['Tumor Volume'].nunique()
# Assemble the resulting series into a single summary dataframe.

#mean = mean.tolist()
#median = median.tolist()
#variance = variance.tolist()
#std_dev = std_dev.tolist()
#sem = sem.tolist()

sum_stats_dict = {'Mean': mean,
                 'Median': median,
                 'Variance': variance,
                 'Standard Deviation': std_dev,
                 'Standard Error or Measurement': sem}
                 #'Drug Regimen': regimen}

summary_stats_df = pd.DataFrame(sum_stats_dict)

summary_stats_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen)

# Using the aggregation method, produce the same summary statistics in a single line
sum_stats_df = combined_df.groupby('Drug Regimen').agg({'Tumor Volume': ['mean', 'median', 'var', 'std', 'sem']})
                                                                        
sum_stats_df
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.

# get count of total timepoints for each drug regimen
timepoint_count = pd.DataFrame(combined_df.groupby(['Drug Regimen']).count().reset_index())

# create df with total timepoints and set the index to drug regimen
timepoint_df = timepoint_count[['Drug Regimen', 'Timepoint']].set_index('Drug Regimen')           
timepoint_df.plot(kind="bar", figsize=(15,3), legend=False)
plt.title('Distribution of Male and Female Mice')
plt.ylabel('Number of Timepoints')
plt.show()
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.

# create list of drug regimens for x axis labels - list of timepoint count for y axis limit
drug_list = combined_df['Drug Regimen'].unique().tolist()
timepoint_list = timepoint_df['Timepoint'].tolist()

# set x asis labels - set size of bar plot - create bar plot chart
x_axis = drug_list
plt.figure(figsize=(15, 3))
plt.bar(x_axis, timepoint_list, color='b', alpha=0.5)

# set x and y limits
plt.xlim(-0.75, len(x_axis))
plt.ylim(0, max(timepoint_list)+10)

# creat title, x axis label, y label
plt.title('Number of Timepoints for Each Drug Regimen')
plt.xlabel('Drug Regimen')
plt.ylabel('Number of Timepoints')
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
gender_df = pd.DataFrame(combined_df.groupby(['Sex']).count()).reset_index()
gender_df = gender_df[['Sex', 'Mouse ID']]
gender_df.plot(kind='pie', y='Mouse ID', labels=gender_df['Sex'], autopct='%1.2f%%', legend=False)
plt.title('Distribution of Male and Female Mice')
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
# create list with total amount of male and female mice
gender_list = combined_df.groupby(['Sex'])['Weight'].count().tolist()


# create pie plot using gender list
plt.pie(gender_list, labels=gender_df['Sex'], colors=['b', 'orange'], autopct='%1.2f%%')

# add title 
plt.title('Distribution of Male and Female Mice')
plt.show()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse:
# find last timepoint for each mouse 
tumor_df = combined_df.loc[combined_df['Timepoint'] == 45]
# set df - sort by drug regimen, mouse id, timepoint
tumor_df = tumor_df.sort_values(['Drug Regimen', 'Mouse ID', 'Timepoint'])
```

```python
# Start by getting the last (greatest) timepoint for each mouse
cap_df = tumor_df[tumor_df['Drug Regimen'].isin(["Capomulin"])] #.reset_index()
cap = cap_df.sort_values(['Tumor Volume'], ascending=True).reset_index()
# final timepoints for capo
cap = cap['Tumor Volume']

# Capo - Calculate the IQR and quantitatively determine if there are any potential outliers. 
quartiles = cap.quantile([.25, .5, .75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of Capomulin is: {lowerq}")
print(f"The upper quartile of Capomulin is: {upperq}")
print(f"The interquartile range of Capomulin is: {iqr}")
print(f"The the median of Capomulin is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```
